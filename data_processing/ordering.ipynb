```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import html
import re
import random
from itertools import permutations
```

```python
url = 'https://en.wikipedia.org/wiki/Timeline_of_the_21st_century'
response = requests.get(url)
response.encoding = 'utf-8'
soup = BeautifulSoup(response.content, 'html.parser')

# Get all the events under the 'mw-parser-output' class
events = soup.find('div', {'class': 'mw-parser-output'}).find_all(['h3', 'li'])

# Initialize a list to store the events, year, and month/day
data = []

# Iterate through each event and extract the date and event text
for event in events:
    text = event.get_text()
    text = html.unescape(text)

    if event.name == 'h3':
        current_year = text.strip('[]=')
        current_year = current_year.split('[')[0]
    elif event.name == 'li':
        if len(text.split(':')) > 1:  # Check if the text contains a date
            date, event_text = text.split(':', 1)
            event_text = event_text.strip()
            month_day = ''

            # Check if there's a '-'
            if '-' in date:
                parts = date.split('-')

                # Both parts have an alphabetic character, indicating month to month
                if any(part.isalpha() for part in parts[0].split()) and any(part.isalpha() for part in parts[1].split()):
                    month_day = parts[0].strip() + " - " + parts[1].strip()

                # Only the first part has an alphabetic character, indicating month-day to day
                elif any(part.isalpha() for part in parts[0].split()):
                    month = ' '.join([word for word in parts[0].split() if word.isalpha()])
                    month_day = month + ' ' + parts[0].split()[0].strip() + " - " + month + ' ' + parts[1].strip()

            elif len(date.split()) > 1:  # Regular date format without '-'
                month_day = ' '.join(date.split()[:2])  # This takes month and day

            data.append([current_year, month_day, event_text])

# Convert the data into a pandas DataFrame
df = pd.DataFrame(data, columns=['Year', 'Date', 'Event'])
df = df.drop_duplicates(subset=['Event'])
```

```python
url = 'https://en.wikipedia.org/wiki/Timeline_of_the_19th_century'
response = requests.get(url)
response.encoding = 'utf-8'
soup = BeautifulSoup(response.content, 'html.parser')

# Get all the events under the 'mw-parser-output' class
events = soup.find('div', {'class': 'mw-parser-output'}).find_all(['h3', 'li'])

# Initialize a list to store the events, year, and month/day
data = []

# Find all the h3 tags
h3_tags = soup.find_all('h3')

for h3 in h3_tags:
    year = h3.get_text().replace('[edit]', '').strip('[]=')
    # Get all the next siblings of h3 until the next h3
    for sibling in h3.find_next_siblings():
        if sibling.name == 'h3':
            break
        if sibling.name == 'ul':
            for li in sibling.find_all('li'):
                event_text = li.get_text()
                # Check if the event_text starts with a year or year range
                if event_text.split()[0].replace('–', '').isdigit():
                    date = event_text.split(':', 1)[0]
                    event_text = event_text.split(':', 1)[1].strip() if ':' in event_text else event_text
                    data.append([year, date, event_text])
                # Check if the event_text starts with a month and day
                elif any(month in event_text for month in ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']):
                    date = event_text.split(' ', 1)[0]
                    event_text = event_text.split(' ', 1)[1].strip()
                    data.append([year, date, event_text])
                else:
                    data.append([year, '', event_text])

# Create a DataFrame
df = pd.DataFrame(data, columns=['Year', 'Date', 'Event'])
```

```python
# time-event dataframe
df = pd.read_csv("time_event.csv")

def generate_order_question(df, n):
    n = max(2, min(n, 6))
    
    available_rows = df[df['Date'] != '']
    
    if len(available_rows) < n:
        return None, None  # Return None if not enough rows to sample
    
    # Randomly select n events without sorting
    sample_df = available_rows.sample(n)
    
    # Shuffle the index for the answer
    shuffled_indices = list(sample_df.index)
    random.shuffle(shuffled_indices)
    
    # Create the question
    question = "Arrange the following events in chronological order: " 
    question += " ".join([f"({i+1}) {sample_df['Event'].loc[idx]}" for i, idx in enumerate(shuffled_indices)])
    
    # The answer is the sorted order of shuffled events based on the DataFrame's order
    sorted_indices = sorted(range(len(shuffled_indices)), key=lambda k: shuffled_indices[k])
    answer = "".join([f"({i+1})" for i in sorted_indices])
    
    return question, answer

def generate_verification_question(df, n, answer_type=None):
    n = max(2, min(n, 5))
    
    available_rows = df[df['Date'] != '']
    
    if len(available_rows) < n:
        return None, None  # Return None if not enough rows to sample
    
    # Randomly select n events and remember their indices in the original DataFrame
    sample_df = available_rows.sample(n)
    original_indices = sample_df.index.tolist()
    
    # Create a list of events
    events = list(sample_df['Event'])
    
    if answer_type is None:
        # Randomly decide whether to generate a "True" or "False" question
        answer_type = random.choice(["True", "False"])
    
    if answer_type == "False":
        # Shuffle events for a "False" question
        random.shuffle(events)
    
    # Determine the correct order of events from the sample_df based on original DataFrame order
    correct_order_events = [events[original_indices.index(idx)] for idx in sorted(original_indices)]
    
    # Find out if the events are in the correct chronological order
    answer = "True" if events == correct_order_events else "False"
    
    # Create the question
    question = "Is the following sequence of events in the correct chronological order? " 
    question += " ".join([f"({i+1}) {event}" for i, event in enumerate(events)])
    
    return question, answer

questions = []
answers = []

for _ in range(15000): 
    q, a = generate_order_question(df, random.randint(2, 6))
    if q and a:
        questions.append(q)
        answers.append(a)

for _ in range(12000):     
    q, a = generate_verification_question(df, random.randint(2, 5))
    if q and a:
        questions.append(q)
        answers.append(a)

# Create DataFrame
df_facts_questions = pd.DataFrame({
    'Question': questions,
    'Answer': answers
})
df_facts_questions = df_facts_questions.drop_duplicates(subset=['Question'])
df_facts_questions['Category'] = 'Facts'
```

```python
def generate_ordering(scenario):
    random_order = random.sample(scenario, len(scenario))
    question = "Arrange the following events in chronological order: " + " ".join(
        [f"({i+1}) {event['Event']}" for i, event in enumerate(random_order)]
    )
    answer = "".join(
        [f"({random_order.index(event) + 1})" for event in scenario]
    )
    return {"question": question, "answer": answer}

def generate_tf(scenario):
    events = scenario["events"]
    event_1, event_2 = random.sample(events, 2)
    if event_1["year"] < event_2["year"]:
        question = f"{event_1['event']} Then {event_2['event']} - True/False?"
        answer = "True"
    else:
        question = f"{event_2['event']} Then {event_1['event']} - True/False?"
        answer = "False"
    return {"question": question, "answer": answer}

def generate_ordering(scenario):
    events = scenario["events"]
    num_events = random.randint(2, len(events))
    sampled_events = random.sample(events, num_events)
    question = "Arrange the following events in chronological order: "
    for idx, ev in enumerate(sampled_events, 1):
        question += f"({idx}) {ev['event']} "
    answer = "".join([f"({i+1})" for i in sorted(range(num_events), key=lambda x: sampled_events[x]['year'])])
    return {"question": question, "answer": answer}

def generate_verification(scenario):
    events = scenario["events"]
    if random.choice([True, False]):  # If true, events are in correct order. 
        question = "Is the following sequence of events in the correct chronological order? "
        for idx, ev in enumerate(events, 1):
            question += f"({idx}) {ev['event']}"        
        answer = "True"
    else:  # If false, shuffle the events randomly.
        shuffled_events = random.sample(events, len(events))
        question = "Is the following sequence of events in the correct chronological order? "
        for idx, ev in enumerate(shuffled_events, 1):
            question += f"({idx}) {ev['event']} "
        answer = "False"
    return {"question": question, "answer": answer}

questions = []
answers = []

repeat_count = 10  # Adjust this as needed to get the desired number of questions

for scenario in scenarios:
    for _ in range(repeat_count):
        tf = generate_tf(scenario)
        order_q = generate_ordering(scenario)
        ver = generate_verification(scenario)

        questions.extend([tf["question"], order_q["question"], ver["question"]])
        answers.extend([tf["answer"], order_q["answer"], ver["answer"]])

# Create a dataframe to store all generated questions and answers
df_real_questions = pd.DataFrame({
    'Question': questions,
    'Answer': answers
})
df_real_questions = df_real_questions.drop_duplicates(subset=['Question'])
df_real_questions['Category'] = 'Commonsense'
```
